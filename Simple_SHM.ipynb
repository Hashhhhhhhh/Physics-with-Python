1. HARMONIC OSCILLATOR:

To plot the different positions of a harmonic oscillator:

In order to simulate the positions of the SHM we need its equation of motion which is x=Acos(wt+Î¦). Initial conditions : x=0 and t=0. The goal is the plot the x vs t graph. We use matplotlib for plotting. In order to plot x(t), we use the time array.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#initializing the parameters:
A=[2,4,6,8]
w=[1,2,3]
phi=0
time=np.linspace(0,10,100)
```

```python
#equation of motion:
for amp in A:  # Iterate through each element in the list A
  for freq in w:  # Iterate through each element in the list w
     x = amp * np.cos(freq * time + phi) # Calculate x using the current amp and
     plt.figure()
     plt.plot(time, x, label=f'A={amp}, w={freq}')
     plt.xlabel('Time (s)')
     plt.ylabel('Position (m)')
     plt.title(f'Simple Harmonic Motion (A={amp}, w={freq})')
     plt.legend()
     plt.grid(True)
     plt.show()
```
